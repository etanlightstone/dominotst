```python
# https://medium.com/@jebaseelanravi96/machine-learning-iris-classification-33aa18a4a983
# Get / work with the data
import numpy as np
import pandas as pd

input_file = "/mnt/data/iris/iris.csv"

iris_df = pd.read_csv(input_file, header = 0)
iris_df.head()
```

```python
iris_df = pd.get_dummies(iris_df, columns=["color"])

```

```python
# DATA PREP AND STORE DATASET

iris_df = iris_df.dropna()
#iris_df = pd.get_dummies(iris_df, columns=["color"])

iris_df.to_csv('iris_ads.csv')
```

```python
# READ THE FINAL ADS

input_file = "iris_ads.csv"

iris_df = pd.read_csv(input_file, header = 0)
iris_df.head()
iris_df.shape
```

```python
# warning 'color' is a categorical calue

iris_dr = pd.read_csv(input_file, header = 0)

feature_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm','color_green','color_pink','color_purple','color_orange','color_red','color_black','color_white']

x = iris_df.loc[:, feature_cols]
y = iris_df.Species


from sklearn.model_selection import GridSearchCV, train_test_split

# here we're splitting the data into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.5)

from sklearn import tree


classifier=tree.DecisionTreeClassifier()

# here we're training
classifier.fit(x_train,y_train)
predictions=classifier.predict(x_test)

# How did we do?
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```
