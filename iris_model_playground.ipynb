```python
# Get / load  the data
import numpy as np
import pandas as pd

input_file = "/mnt/data/dat/iris.csv"

iris_df = pd.read_csv(input_file, header = 0)
iris_df.head()
```

```python
# preparing the data
#
# get rid of fields with null values
iris_df.isnull()
iris_df = iris_df.dropna()

# Color is a category with text values, to make this work it needs to be encoded as values:
iris_df = pd.get_dummies(iris_df, columns=["color"])

iris_df.head()
```

```python

# warning 'color' is a categorical calue
feature_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm','color_green','color_pink','color_purple','color_orange','color_red','color_black','color_white']

x = iris_df.loc[:, feature_cols]
# species is what we are trying to predict
y = iris_df.Species


from sklearn.model_selection import GridSearchCV, train_test_split

# here we're splitting the data into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.5)

from sklearn import tree


classifier=tree.DecisionTreeClassifier()

# here we're training
classifier.fit(x_train,y_train)
predictions=classifier.predict(x_test)

# How did we do?
from sklearn.metrics import accuracy_score

accuracy_score(y_test,predictions)


print(accuracy_score(y_test,predictions))
```
